Intro to Data Science:
* Data science is all about using data to solve problems. The problem could be decision making such as identifying which email is spam and which is not. 
Different steps involved in data science:
    1. Define Problem Statement
    2. Data Collection
    3. Data Quality Check and Remediation
    4. Exploratory Data Analysis
    5. Data Modelling
    6. Data Communication

Types of Data:

        a. Nominal
            e.g. Male/Female
        b. Ordinal

    2. Quantitative
        a. Ratio
            e.g. heights of a person
        b. Interval

Imputation:
    Imputation is the process of replacing unknown values with the best guess. Usually, we use
    the statistical parameters such as mean, median or mode to impute the missing values.
    When to use which parameter is decided by the parameters ability to provide the best
    central location for the data
    1.Mean
    2.Median
    3.Mode

Confirmatory Data Analysis (CDA):
        In statistics, a hypothesis test calculates some quantity under a given assumption. The result of
    the test allows us to interpret whether the given assumption holds or not. The given assumption
    of a statistical test is called the null hypothesis, or H0. Whereas a contradiction to the null
    hypothesis is called the alternative hypothesis, HA.
