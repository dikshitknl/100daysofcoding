```python
import numpy as np
```

```python
# define input features:
input_features = np.array([[0,0],[0,1],[1,0],[1,1]])

# target output, reshape into vector:
target_output = np.array([[0,1,1,1]]).reshape(4,1)

weights = np.array([[0.1],[0.2]])

bias = 0.3

learning_rate = 0.05

# sigmoid function:
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def der_sig(x):
    x = sigmoid(x)
    return x * (1 - x)
```

```python
#main logic:
for epoch in range(10000):
    inputs = input_features
    
    #feed forward input 
    # dot product of weight and input features and add bias to it:
    in_o = np.dot(inputs, weights) + bias
    
    #feed forward output:
    out_o = sigmoid(in_o)
    
    #back propagation:
    #error calculation:
    error = out_o - target_output
    
    #finding derivative:
    derror_douto = error
    douto_dino = der_sig(out_o)
    
    #multiplying indv derivative:
    deriv = derror_douto * douto_dino
    
    # transpose of input_features
    inputs = input_features.T
    deriv_final = np.dot(inputs, deriv)
    
    #update weight:
    weights -= learning_rate * deriv_final
    
    print(error.sum())
    #updating the bias weight value:
    for i in deriv:
        bias -= learning_rate * i
```
