```python
from random import randint
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

m = 80
data = []
for i in range(m):
    width = randint(1,25)
    w1 = randint(0,12)
    data.append(randint(i+w1, i+w1+width))
```

```python
X = list(range(m))
Y = data

plt.scatter(X,Y,color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary Prediction Curves')
plt.show()
```

```python
# mean of our inputs and outputs
x_mean = np.average(X)
y_mean = np.average(Y)

#total number of values
n = len(X)

# using the formula to calculate the b1 and b0
numerator = 0
denominator = 0
for i in range(n):
    numerator += (X[i] - x_mean) * (Y[i] - y_mean)
    denominator += (X[i] - x_mean) ** 2
    
b1 = numerator / denominator
b0 = y_mean - (b1 * x_mean)

#printing the coefficient

data1 = []
for i in range(m):
    data1.append(b0 + b1 * i)

X = list(range(m))
Y = data

plt.scatter(X,Y,color='blue')
plt.plot(X,data1,color='green')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary Prediction Curves')



def predict(x):
    return (b0 + b1 * x)
x = int(input("Enter value you want to predict: "))
y = int(np.ceil(predict(x)))
plt.plot(range(0,x+1), [y]*(x+1), color='red')
plt.plot([x]*(y+1), range(0,y+1), color='red')

print('Actual: ', data[x])
print('Predicted: ', y)

reg = LinearRegression()
reg.fit(np.array(X).reshape(-1,1),np.array(Y).reshape(-1,1))
print(x, 33333333333333)
y = reg.predict([[x]])

plt.plot(x, y, 'g','x')
plt.show()
print('Actual: ', data[x])
print('Predicted: ', y)
```
