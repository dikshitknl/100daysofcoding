# ARIMA model for time series analysis

```python
from pandas import read_csv
import pandas as pd
from matplotlib import pyplot
from datetime import datetime
from pandas.plotting import autocorrelation_plot
from random import randint

file_name = 'supermarket_sales - Sheet1.csv'

def parser(date_in_str):
    dates, time = date_in_str.split('T') 
    time = time[0:2] + ':' + time[2:] + ':00'
    dates = dates[-2:]+'/'+dates[-4:-2]+'/'+dates[2:4]
    datestime = dates + ' ' + time
    return datetime.strptime(datestime, '%d/%m/%y %H:%M:%S')
```
